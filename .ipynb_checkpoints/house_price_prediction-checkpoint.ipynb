## Linear Rigression, Ridge and Lasso

```python
#house pricing dataset
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = fetch_california_housing(as_frame = True)
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
#independant features
dataset = pd.DataFrame(df.data , columns=df.feature_names)
dataset["target"] = df.target
```

```python
print(dataset.head())
```

```python
dataset.head()
```

```python
#dependant feature
dataset['Price']=df.target
```

```python
dataset.head()
```

```python
#dividing the dataset into dependant and dependant features
X=dataset.iloc[:,:-1] #independant features
y=dataset.iloc[:,-1] #dependant features
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.33,random_state=42)
```

## LINEAR RIGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
mse=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
##lin_reg.predict()
```

## RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]
}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=8)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor)
```

```python
ridge_regressor.best_score_
```

```python
print(ridge_regressor.best_score_)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

## LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]
}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=10)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred=lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score

r2_score1=r2_score(y_pred,y_test)
```

```python
print(r2_score1)
```

```python
y_pred=ridge_regressor.predict(X_test)
from sklearn.metrics import r2_score

r2_score1=r2_score(y_pred,y_test)
```

```python
print(r2_score1)
```

```python
y_pred=lin_reg.predict(X_test)
from sklearn.metrics import r2_score

r2_score1=r2_score(y_pred,y_test)
```

```python
print(r2_score1)
```

## LOGISTIC REG

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
# independent features
X=pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
# dependent features
y=pd.DataFrame(df['target'],columns=["Target"])
y
```

```python
# checking if column y is balanced or imbalanced
y['Target'].value_counts()
# and as the below results this is balanced dataset
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
X,y,test_size=0.33,random_state=42)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
print(model)
```

```python
print(model.best_score_)
```

```python
print(model.best_params_)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
